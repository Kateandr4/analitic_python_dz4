## Условие 3: 3 задача
Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.
Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
corr_matrix = df.corr()
corr_matrix = np.round(corr_matrix, 2)
corr_matrix 
```

```python
corr_m = df.corr().round(1)
corr_m[np.abs(corr_m)<0.3] = 0
corr_m
```

```python
plt.figure(figsize=(12,8))
sns.set(font_scale=1.4)
sns.heatmap(corr_m, annot=True, linewidths =0.5, cmap='coolwarm')
```

По степени корреляции видим, что сильную прямую зависимость с ценой показывают следующие факторы:
Общая площадь дома (0,7);
Оценка дома (0.7);
Средняя площадь дома у ближайших 15 соседей участка (0.6);
Площадь дома без учета подвала (0.6);

в меньшей степени зависимость от таких факторов как:
Кол-во ванных комнат (0.5);
Наличие вида (0.4).

```python
plt.figure(figsize=(8,6))
sns.boxplot(y=df.grade, x=df.price/1000, whis=1.5, orient='h')
plt.title('Зависимость цены от оценки дома')
plt.xlabel('Цена')
plt.ylabel('Оценка');
```

Оценка дома значительно зависит от цены для домов стоимостю более 5000 тыс еввро
